```python
# Quesiton 13
```

```python
df_data = df.copy()  # to restore the original dataset 
df_data.head()
```

```python
df_data.drop(['date' , 'lights'] , axis=1 , inplace=True)
```

```python
cols_df_data=df_data.columns  # copying the column names for later references
cols_df_data
```

```python
df_data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df_data=scaler.fit_transform(df_data)
```

```python
# converting again into dataframe
df_data=pd.DataFrame(df_data , columns = cols_df_data)
df_data.head()
```

```python
# Train_test_split
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(df_data.drop('Appliances' , axis=1) , df_data['Appliances'] , 
                                                      test_size=0.3 , random_state=42)
```

```python
X_train.shape , X_test.shape
```

```python
y_train.shape , y_test.shape
```

```python
# Model Multiple Linear Regression
model_multiple_linear_regression = LinearRegression().fit(X_train , y_train)
```

```python
predictions_test = model_multiple_linear_regression.predict(X_test)
```

```python
from sklearn import metrics
value_mae = metrics.mean_absolute_error(y_test , predictions_test)
np.round(value_mae,2)
```

```python
# Quesiton 14
```

```python
value_rss = ((y_test-predictions_test)**2).sum()
np.round(value_rss,2)
```

```python
# Quesiton 15
```

```python
value_rss = metrics.mean_squared_error(y_test , predictions_test , squared=False)
np.round(value_rss,3)
```

```python
# Quesiton 16
```

```python
# RSS (Residual Sum of Squares)
value_r2_score = metrics.r2_score(y_test , predictions_test)
np.round(value_r2_score , 2)
```

```python
from sklearn.linear_model import Ridge
model_ridge_regression = Ridge(alpha=0.4).fit(X_train , y_train)
```
